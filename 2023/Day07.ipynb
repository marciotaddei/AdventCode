```python
import numpy as np
from collections import Counter
import numpy.lib.recfunctions as nlr
```

```python
# hands = np.loadtxt("exampleinput7.dat", dtype=[('hand', 'U5')], usecols=0)
# bids = np.loadtxt("exampleinput7.dat", dtype=[('bid', int)], usecols=1)
data = np.loadtxt("input7.dat", dtype=[('hand', 'U5'), ('bid', int)])
orddict = {ch: '0123456789TJQKA'.find(ch) for ch in '0123456789TJQKA'}
numbered = np.array([tuple([orddict[ch] for ch in hand]) for hand in data['hand']], dtype=[('num'+str(j), int) for j in range(5)]) 
types = [sorted(Counter(hand+'0').values(), reverse=True) for hand in data['hand']] #'0' added for the counts of five of a find to have a second element
types = np.array([type[0]*10+type[1] for type in types], dtype=[('type', int)])
data = nlr.merge_arrays([data, numbered, types], flatten=True)
data.sort(order=['type']+['num'+str(j) for j in range(5)])
bidsum = 0
for i in range(data.shape[0]):
    bidsum = bidsum + (i+1)*data[i]["bid"]
bidsum
```

```python
data = np.loadtxt("input7.dat", dtype=[('hand', 'U5'), ('bid', int)])
orddict = {ch: '0J23456789TQKA'.find(ch) for ch in '0123456789TJQKA'}
numbered = np.array([tuple([orddict[ch] for ch in hand]) for hand in data['hand']], dtype=[('num'+str(j), int) for j in range(5)]) 
def type_of(hand):
    n_J = Counter(hand)['J']
    out = sorted(Counter(hand+'0').values(), reverse=True) #'0' added for the counts of five of a find to have a second element
    if n_J>0 and n_J<5:
        out.remove(n_J)
        out[0] = out[0] + n_J
    return out[0]*10+out[1]
types = np.array([type_of(hand) for hand in data['hand']], dtype=[('type', int)])
data = nlr.merge_arrays([data, numbered, types], flatten=True)
data.sort(order=['type']+['num'+str(j) for j in range(5)])
bidsum = 0
for i in range(data.shape[0]):
    bidsum = bidsum + (i+1)*data[i]["bid"]
bidsum
```
