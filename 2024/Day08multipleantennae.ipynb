```python
import matplotlib.pyplot as plt
import numpy as np
np.set_printoptions(linewidth=200, threshold=200)
```

```python
themap = np.genfromtxt('input08.txt', dtype=str, delimiter=1)
antinodes = np.zeros_like(themap, dtype=bool)
chars, maplen = [char for char in np.unique(themap) if char !='.'], themap.shape[0]
for char in chars:
    locs = np.argwhere(themap == char)
    for i in range(len(locs)):
        for j in range(i+1,len(locs)):
                vec = 2*locs[j]-locs[i]
                if (vec>=0).all() and (vec<maplen).all(): #using that map is square
                    antinodes[tuple(vec)] = True
                vec = 2*locs[i]-locs[j]
                if (vec>=0).all() and (vec<maplen).all(): #using that map is square
                    antinodes[tuple(vec)] = True
plt.matshow(antinodes)
antinodes.sum()
```

```python
themap = np.genfromtxt('input08.txt', dtype=str, delimiter=1)
antinodes = np.zeros_like(themap, dtype=bool)
chars, maplen = [char for char in np.unique(themap) if char !='.'], themap.shape[0]
for char in chars:
    locs = np.argwhere(themap == char)
    for i in range(len(locs)):
        for j in range(i+1,len(locs)):
                vec = locs[j]
                while (vec>=0).all() and (vec<maplen).all(): #using that map is square
                    antinodes[tuple(vec)] = True
                    vec = vec + locs[j]-locs[i]
                vec = locs[i]
                while (vec>=0).all() and (vec<maplen).all(): #using that map is square
                    antinodes[tuple(vec)] = True
                    vec = vec + locs[i]-locs[j]
plt.matshow(antinodes)
antinodes.sum()
```
