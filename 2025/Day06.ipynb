Part One

```python
import numpy as np
np.set_printoptions(linewidth=np.inf)

content = np.genfromtxt('input06.txt', dtype=str)
numbers = content[:-1].astype(int)
operations = content[-1]

result = 0
for i in range(len(operations)):
    result += np.sum(numbers[:,i]) if operations[i]=="+" else np.prod(numbers[:,i])
result
```

```python
#this avoids the for loop, but calculates the sum and product for every column, not just where needed
sums, prods = np.sum(numbers, axis=0), np.prod(numbers, axis=0)
is_sum = operations == "+"
np.sum(is_sum*sums + (~is_sum)*prods)
```

Part Two

```python
import numpy as np
np.set_printoptions(linewidth=np.inf)

content = np.genfromtxt('input06.txt', dtype=str, delimiter=1)
operations = list(content[-1])
while ' ' in operations: operations.remove(' ')

numbers = np.apply_along_axis(lambda row: ''.join(row), axis=1, arr=content[:-1].T)
for i in range(20):
    if ' '*i in numbers: empty_str = ' '*i; break
numbers = np.split(numbers, np.where(numbers==empty_str)[0])
numbers = [[int(n) for n in group if n!=empty_str] for group in numbers]

result = 0
for i in range(len(operations)):
    result += np.sum(numbers[i]) if operations[i]=="+" else np.prod(numbers[i])
result
```
