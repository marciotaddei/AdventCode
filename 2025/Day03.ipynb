Part One

```python
with open("input03.txt", "r") as f:
    ratings = [tuple(int(n) for n in row.strip("\n")) for row in f.readlines()]
```

```python
joltage = 0
for row in ratings:
    n0 = max(row[:-1]) #largest digit other than last
    n0_idx = row.index(n0)
    n1 = max(row[n0_idx+1:]) #largest subsequent digit
    # print(n0, n0_idx, n1, 10*n0+n1)
    joltage += 10*n0+n1
joltage
```

Part Two

```python
with open("input03.txt", "r") as f:
    ratings = [row.strip("\n")+' ' for row in f.readlines()]
```

```python
#A little gambiarra to avoid slicing [:-0] instead of [:]. Every row has a last, useless character ' ', then slicing goes from [:-12] to [:-1]
joltage = 0
for row in ratings:
    sequence = ''
    for i in range(12):
        n = max(row[:-12+i]) #gambiarra here avoids an if else statement here
        n_idx = row.index(n)
        row = row[n_idx+1:]
        sequence = sequence + n
    joltage += int(sequence)
    # print(sequence)
joltage
```
