```python
def bisect_find(the_list, num):
    """Custom bisect function to find index of inserting num in the_list.
    Assumes sorted the_list, if equal, bisect_right"""
    min, max = 0, len(the_list)
    while max-min>1:
        ind = (min+max)//2
        if num>=the_list[ind]: min=ind
        else: max=ind
    if min==0 and num<the_list[0]: return 0
    return max
```

```python
#import bisect custom bisect

with open("input05.txt", 'r') as f:
    the_input = f.readlines()
breakpoint = the_input.index("\n")
fresh, IDs = [tuple(int(n) for n in x.strip("\n").split('-')) for x in the_input[:breakpoint]], [int(x) for x in the_input[breakpoint+1:]]

fresh = sorted([(i,j+1) for i,j in fresh]) #range-end exclusive
for i in range(len(fresh)-1):
    if fresh[i][1]>=fresh[i+1][0]: #merge ranges
        fresh[i+1] = (fresh[i][0],max(fresh[i][1],fresh[i+1][1])) #max because one interval may just fully include the other
        fresh[i] = ()
while () in fresh: fresh.remove(())
fresh = [n for pair in fresh for n in pair]
n_fresh_IDs = 0
for ID in IDs:
    pos = bisect_find(fresh, ID)#bisect.bisect(fresh, ID)
    if pos%2: n_fresh_IDs+=1
    # print(pos, ID, fresh[pos-1:pos+1], n_fresh_IDs)
n_fresh_IDs
```
