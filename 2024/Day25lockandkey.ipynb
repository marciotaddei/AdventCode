```python
import numpy as np
```

```python
full_array = np.genfromtxt("input25.txt", comments=None, delimiter=1, dtype=str)
```

```python
locks, keys = set(), set()
for i in range(0,full_array.shape[0],7):
    subarray = full_array[i:i+7]
    if subarray[0,0] == "#": #lock
        thisone = tuple(np.count_nonzero(subarray[:,j]=="#")-1 for j in range(subarray.shape[1]))
        locks.add(thisone)
    elif subarray[-1,0] == "#": #key
        thisone = tuple(np.count_nonzero(subarray[:,j]=="#")-1 for j in range(subarray.shape[1]))
        keys.add(thisone)
    else: raise ValueError("Something's wrong with the array at i=", i)
del full_array
# locks, keys
```

```python
ct=0
for k in keys:
    for l in locks:
        if not set(map(sum, zip(k,l))).intersection({6,7,8,9,10}):
            ct +=1
ct
```
