```python
from multiset import Multiset as mset

inputexample = "029A 980A 179A 456A 379A".split()
theinput = "083A 935A 964A 149A 789A".split()

rev_num_pad = {"A":(0,2), "0":(0,1)} #rows growing upward!!!!
rev_num_pad.update({str(n): ( (n+2)//3, (n+2)%3) for n in range(1,10)})
rev_arrow_pad = {"A":(1,2), "^":(1,1), "<":(0,0), "v":(0,1), ">":(0,2)} #rows growing upward!!!!
```

```python
def commands_numpad(code :str):
    commands = ""
    current = rev_num_pad["A"]
    for ch in code:
        target = rev_num_pad[ch]
        ver_comm = "^"*(target[0]-current[0]) if target[0]>current[0] else "v"*(current[0]-target[0])
        hor_comm = ">"*(target[1]-current[1]) if target[1]>current[1] else "<"*(current[1]-target[1])
        if "<" in hor_comm and (current[0]!=0 or target[1]!=0):
            commands += hor_comm + ver_comm
        elif current[1]==0 and target[0]==0:#avoid empty square
                commands += hor_comm + ver_comm 
        else: commands += ver_comm + hor_comm
        commands += "A"
        current = target
    return commands

#improved order also here
def commands_arrowpad(code :str):
    commands = ""
    current = rev_arrow_pad["A"]
    for ch in code:
        target = rev_arrow_pad[ch]
        ver_comm = "^"*(target[0]-current[0]) if target[0]>current[0] else "v"*(current[0]-target[0])
        hor_comm = ">"*(target[1]-current[1]) if target[1]>current[1] else "<"*(current[1]-target[1])
        if ch == "<": #avoid empty square
            commands += ver_comm + hor_comm
        elif current[1] == 0: #avoid empty square
            commands += hor_comm + ver_comm
        elif "<" in hor_comm: commands += hor_comm + ver_comm
        else: commands += ver_comm + hor_comm
        commands += "A"
        current = target
    return commands

def comms_as_mset(arrowcode: mset):
    out = mset()
    for chunk, mult in arrowcode.items():
        for spl in [s+"A" for s in commands_arrowpad(chunk).split("A")[:-1]]:
            out.add(spl, mult)
    return out
```

```python
# We need to break it into chunks ending on A. There are many of those, but very few variation (only like 8 different ones).
# Just keep counting the multiplicity of each unique one. For that, multisets!
complexity = 0
for code in theinput:
    arrowcode = mset([s+"A" for s in commands_numpad(code).split("A")[:-1]])
    for i in range(25): arrowcode = comms_as_mset(arrowcode)
    length = sum(len(k)*v for k,v in arrowcode.items())
    print(code, length)
    complexity += int(code[:-1]) * length
complexity   
```
