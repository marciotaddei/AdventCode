```python
filename = "input8.dat"
with open(filename, "r") as f:
    LRseries = f.readline()[:-1]
    s = re.sub("[=,()]", "", f.read()) #because np.loadtxt doesn't support different delimiters at once
coords = np.loadtxt(io.StringIO(s), str, skiprows=1)

step, looped_step, reached = 0,0,set()
positions = {pos for pos in coords[:,0] if pos[2]=='A'} #initial positions
while positions!=set():
    if LRseries[looped_step]=="L":
        LR = 1
    elif LRseries[looped_step]=="R":
        LR = 2
    else:
        raise IndexError("LR series over, perhaps missing loop around?")
    # pos = coords[np.argwhere(coords[:,0]==pos)][0,0][LR]
    positions = {coords[np.argwhere(coords[:,0]==pos)][0,0][LR] for pos in positions}
    step = step + 1
    looped_step = step % len(LRseries)
    reached = reached.union({(pos, step) for pos in positions if pos[2]=="Z"}) #could've kept just the steps
    positions = {pos for pos in positions if pos[2]!="Z"}
    # print(step, looped_step, LR, positions)
reached = np.array([[p, s] for p,s in reached])
reached, math.lcm(*(reached[:,1].astype(int)))
```
