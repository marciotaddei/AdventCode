Part One

```python
import time
t0 = time.time()

with open("input11.txt") as f:
    data = f.readlines()
data = {s.strip("\n").split(":")[0]: {to for to in s.strip("\n").split(":")[1].strip().split()}  for s in data}

partial_paths, full_paths = [['you']], []
while partial_paths:
    current = partial_paths.pop(0)
    input = current[-1]
    for output in data[input]:
        if output == 'out': full_paths.append(current+[output])
        else: partial_paths.append(current+[output])
print(f"A total of {len(full_paths)} paths")
full_paths if len(full_paths) < 10 else None
print(f"Solved with recurson in {time.time()-t0:4f} seconds")
```

Part Two

```python
import time
# from functools import cache

t0 = time.time()
with open("input11.txt") as f:
    data = f.readlines()
data = {s.strip("\n").split(":")[0]: {to for to in s.strip("\n").split(":")[1].strip().split()}  for s in data}

#recursive function
cache = {}
# @cache
def n_paths(start, finish, excluded=()):
    if (start, finish) in cache: return cache[(start, finish)]
    if start==finish: return 1
    if start in excluded: return 0
    # return sum(n_paths(node, finish, excluded) for node in data[start]) #with @cache
    output = sum(n_paths(node, finish, excluded) for node in data[start])
    cache[(start, finish)] = output
    return output

n_paths_svr_fft = n_paths('svr', 'fft', ('dac', 'out'))
n_paths_svr_dac = n_paths('svr', 'dac', ('fft', 'out'))
n_paths_fft_dac = n_paths('fft', 'dac', ('out'))
n_paths_dac_fft = n_paths('dac', 'fft', ('out'))
n_paths_fft_out = n_paths('fft', 'out', ('dac'))
n_paths_dac_out = n_paths('dac', 'out', ('fft'))

print(n_paths_svr_dac*n_paths_dac_fft*n_paths_fft_out + n_paths_svr_fft*n_paths_fft_dac*n_paths_dac_out)
print(f"Solved with recurson in {time.time()-t0:4f} seconds")
```

Recursive functions are AWESOME! With a cache, of course. Both parts done in no time (<0.1s)
