This one super simple if you know some basic linear-algebraic methods to solve systems of linear equations

Part I

```python
config = []
with open("input13.txt", "r") as file:
    data = file.read().splitlines()
for i in range(0, len(data),4):
    buttonA = data[i].split(":")[1].split(",")
    buttonB = data[i+1].split(":")[1].split(",")
    prize   = data[i+2].split(":")[1].split(",")
    config.append((int(buttonA[0].strip(" XY+")), int(buttonA[1].strip(" XY+")), 
                   int(buttonB[0].strip(" XY+")), int(buttonB[1].strip(" XY+")), 
                   int(prize[0].strip(" XY=")), int(prize[1].strip(" XY="))))
```

```python
tokens = 0
for machine in config:
    det = (machine[0]*machine[3] - machine[2]*machine[1]) # never zero, tested
    detA = (machine[4]*machine[3] - machine[5]*machine[2])
    detB = (machine[0]*machine[5] - machine[1]*machine[4])
    n, m = detA/det, detB/det #simple Cramer's rule
    if n.is_integer() and m.is_integer():
        # print(n, m)
        tokens += int(3*n+m)
tokens
```

Part II

```python
config = []
with open("input13.txt", "r") as file:
    data = file.read().splitlines()
for i in range(0, len(data),4):
    buttonA = data[i].split(":")[1].split(",")
    buttonB = data[i+1].split(":")[1].split(",")
    prize   = data[i+2].split(":")[1].split(",")
    config.append((int(buttonA[0].strip(" XY+")), int(buttonA[1].strip(" XY+")), 
                   int(buttonB[0].strip(" XY+")), int(buttonB[1].strip(" XY+")), 
                   int(prize[0].strip(" XY="))+10000000000000, int(prize[1].strip(" XY="))+10000000000000))
```

```python
tokens = 0
for machine in config:
    det = (machine[0]*machine[3] - machine[2]*machine[1]) #assumed not zero (tested so)
    detA = (machine[4]*machine[3] - machine[5]*machine[2])
    detB = (machine[0]*machine[5] - machine[1]*machine[4])
    n, m = detA/det, detB/det #simple Cramer's rule
    if n.is_integer() and m.is_integer():
        # print(n, m, machine)
        tokens += int(3*n+m)
tokens
```
