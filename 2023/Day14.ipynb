```python
import numpy as np
```

```python
def tilt_west(grid):
    for i in range(grid.shape[0]):
        new_row = []
        indices = np.where(grid[i]=="#")[0]
        subrows = [grid[i][beg:end] for beg,end in zip(np.concatenate(([0],indices+1)), np.concatenate((indices+1, [grid.shape[1]])))]
        for subrow in subrows:
            new_row += sorted(subrow, reverse=True)
        grid[i] = np.array(new_row)
    return grid

def load_north(grid):
    load = 0
    for i in range(grid.shape[0]):
        load += np.count_nonzero(grid[i]=="O")*(grid.shape[0]-i)
    return load

def tilt_cycle(grid):
    grid = tilt_west(grid.T).T #tilt north
    grid = tilt_west(grid)
    grid = tilt_west(grid.T[:,::-1])[:,::-1].T #tilt south
    grid = tilt_west(grid[:,::-1])[:,::-1] #tilt east
    return grid
```

```python
grid = np.genfromtxt("input14.dat", str, delimiter=1, comments=None)
grid = tilt_west(grid.T).T #tilt north
load_north(grid)
```

```python
grid = np.genfromtxt("input14.dat", str, delimiter=1, comments=None)
n_total = 1000000000
storage = []
for i in range(n_total):
    if grid.tobytes() in storage: #storing as array raises problems
        circle_begin = storage.index(grid.tobytes())
        circle_len = i - circle_begin
        break
    storage.append(grid.tobytes())
    grid = tilt_cycle(grid)
n_equiv = (n_total - circle_begin) % circle_len +circle_begin
final_grid = storage[n_equiv]
final_grid = np.frombuffer(final_grid, dtype="S4").astype(str).reshape(grid.shape)
load_north(final_grid)
```
