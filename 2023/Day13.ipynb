```python
import numpy as np
```

```python
def reflection_row(grid, sauf=None): #sauf relevant for Part 2
    for j in [j for j in range(1,grid.shape[0]) if j!=sauf]:
        low_end = max(0, 2*j - (grid.shape[0]))
        high_end = min(2*j, grid.shape[0])
        if (grid[low_end:j]==grid[j:high_end][::-1]).all():
            return j
    return 0
```

```python
with open("input13.dat", "r") as f:
    lines = f.readlines() + ["\n"]
the_sum = 0
while lines !=[]:
    idx = lines.index('\n')
    grid = np.array([[ch for ch in line.removesuffix("\n")] for line in lines[:idx]])
    the_sum = the_sum + reflection_row(grid.T) #cols
    the_sum = the_sum + 100*reflection_row(grid) #rows
    lines = lines[idx+1:]
the_sum
```

```python
with open("input13.dat", "r") as f:
    lines = f.readlines() + ["\n"]
the_sum = 0
while lines !=[]:
    idx = lines.index('\n')
    grid = np.array([[ch for ch in line.removesuffix("\n")] for line in lines[:idx]])
    old_reflect = (reflection_row(grid), reflection_row(grid.T))
    breakout = False
    for i in range(grid.shape[0]):
        if breakout:
            break
        for j in range(grid.shape[1]):
            grid[i,j] = "." if grid[i,j]=="#" else "#"
            new_reflect = (reflection_row(grid, sauf=old_reflect[0]), reflection_row(grid.T, sauf=old_reflect[1]))
            if new_reflect==(0,0):
                grid[i,j] = "." if grid[i,j]=="#" else "#" #undo change
            else:
                the_sum = the_sum +     new_reflect[1] #cols
                the_sum = the_sum + 100*new_reflect[0] #rows
                breakout = True
                break
    lines = lines[idx+1:]
the_sum
```
