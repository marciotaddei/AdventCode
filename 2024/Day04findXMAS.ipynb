```python
import numpy as np

array = np.genfromtxt("input04.txt", dtype=str, delimiter=1)
array = np.hstack( ([[0]]*(array.shape[1]+2), np.vstack(([0]*(array.shape[0]), array, [0]*(array.shape[0]))), [[0]]*(array.shape[1]+2)))
```

```python
xmas_subarray = np.array(['X', 'M', 'A', 'S'])
ct = 0
for i in range(array.shape[0]):
    for j in range(array.shape[1]):
        if array[i,j] != "X": continue
        
        if np.array_equal(array[i, j:j+4], xmas_subarray): ct +=1
        if np.array_equal(array[i, j:j-4:-1], xmas_subarray): ct +=1
        if np.array_equal(array[i:i+4, j], xmas_subarray): ct +=1
        if np.array_equal(array[i:i-4:-1, j], xmas_subarray): ct +=1

        if np.array_equal(array[i:i+4, j:j+4].diagonal(), xmas_subarray): ct +=1
        if np.array_equal(array[i:i+4, j:j-4:-1].diagonal(), xmas_subarray): ct +=1
        if np.array_equal(array[i:i-4:-1, j:j+4].diagonal(), xmas_subarray): ct +=1
        if np.array_equal(array[i:i-4:-1, j:j-4:-1].diagonal(), xmas_subarray): ct +=1
ct
```
