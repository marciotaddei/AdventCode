```python
equations = []
with open("input07.txt", "r") as file:
    line = file.readline()
    i = 0 
    while line != "":
        equations.append([int(line.split(":")[0])] + [int(x) for x in line.split(":")[1].split()])
        line = file.readline()
        i +=1
if not i == len(equations): print("error")
```

```python
def verify_eq(lhs: int, terms: list):
    totals = {terms[0]}
    for i in range(1, len(terms)):
        totals = {t + terms[i] for t in totals} | {t*terms[i] for t in totals}
        totals = {t for t in totals if t <= lhs}
    return lhs in totals
```

```python
sum(eq[0] for eq in equations if verify_eq(eq[0], eq[1:]))
```

By using the lhs as dictionary key, some entries with coinciding lhs had been overwritten: "20665830406676 is too low"

```python
def verify_eq_plus(lhs: int, terms: list):
    if verify_eq(lhs, terms): return True #makes it ~20% faster, may be removed
    totals = {terms[0]}
    for term in terms[1:]:
        totals = {t + term for t in totals} | {t*term for t in totals} | {int(str(t)+str(term)) for t in totals}
        totals = {t for t in totals if t <= lhs}
    return lhs in totals
```

```python
sum(eq[0] for eq in equations if verify_eq_plus(eq[0], eq[1:]))
```
