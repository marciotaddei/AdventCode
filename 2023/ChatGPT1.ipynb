```python
def calculate_calibration_sum(lines):
    calibration_sum = 0
    for line in lines:
        # Remove non-digit characters and find the first and last digits
        digits = [char for char in line if char.isdigit()]
        if digits:
            calibration_value = int(digits[0] + digits[-1])
            calibration_sum += calibration_value
    return calibration_sum
```

```python
# Example calibration document
calibration_document = [
    "1abc2",
    "pqr3stu8vwx",
    "a1b2c3d4e5f",
    "treb7uchet"
]

# Calculate and print the sum of calibration values
result = calculate_calibration_sum(calibration_document)
print("Sum of calibration values:", result)
```

```python
with open("input1.dat", "r") as f:
    lines = f.readlines()
result = calculate_calibration_sum(lines)
print("Sum of calibration values:", result)
```
