```python
doc = """L68 L30 R48 L5 R60 L55 L1 L99 R14 L82""".split()

with open("input01.txt", "r") as f:
    doc = [s.strip() for s in f.readlines()]

pos, prevzero, zeroct = 50, False, 0
for instruction in doc:
    pos = pos + (1 if instruction[0]=="R" else -1)*int(instruction[1:])
    zeroct+= abs(pos//100)
    if pos<0:
        if prevzero: zeroct-=1
        if pos%100==0: zeroct+=1
    if pos == 0: zeroct+=1
    # print(pos, pos%100, zeroct)
    pos = pos%100
    prevzero = not pos
zeroct
```
