```python
def check_safe(l: list):
    sign = l[1]-l[0]
    if sign == 0: return False
    sign = int(sign/abs(sign))
    
    for i in range(1,len(l)):
        signed_diff = (l[i]-l[i-1])*sign
        if signed_diff <= 0 or signed_diff > 3:
            return False
    return True
```

```python
count=0
with open("./input02.txt", "r") as file:
    while True: 
        line = file.readline().split()
        if line == []: break
        count += check_safe([int(x) for x in line])
count
```

```python
def check_safe_dampener(l: list):
    sign = l[1]-l[0]
    if sign == 0: return check_safe(l[1:])
    sign = int(sign/abs(sign))
    
    for i in range(1,len(l)):
        signed_diff = (l[i]-l[i-1])*sign
        if signed_diff <= 0 or signed_diff > 3:
            return check_safe(l[:i-1]+l[i:]) or check_safe(l[:i]+l[i+1:]) or (check_safe(l[1:]) if i==2 else 0)
            #the last boolean just for when the first sign is wrong, detected at i=2 and corrected removing the first element
    return True
```

```python
count=0
with open("./input02.txt", "r") as file:
    while True: 
        line = file.readline().split()
        if line == []: break
        count += check_safe_dampener([int(x) for x in line])
count
```
