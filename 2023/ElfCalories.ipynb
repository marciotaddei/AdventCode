```python
with open("./inputCalories.dat", "r") as f:
    lines = f.readlines()
lines.append("\n")
cal_list = []
calories = 0
for line in lines:
    if line == "\n":
        cal_list.append(calories)
        calories = 0
    else:
        calories = calories + int(line)
cal_list = sorted(cal_list)
cal_list[-1], cal_list[-1]+cal_list[-2]+cal_list[-3]
```
