```python
grid = np.genfromtxt("input11.dat", str, delimiter=1, comments=None)
#marking values to be expanded
empty_rows = set()
for i in range(len(grid)):
    if set(grid[i])=={"."}:
        empty_rows.add(i)
empty_cols = set()
for i in range(len(grid.T)):
    if set(grid.T[i])=={"."}:
        empty_cols.add(i)

#measuring distances
def distance(p1, p2):
    distance = abs(p1-p2).sum()
    for row_n in empty_rows:
        if (row_n > p1[0] and row_n <p2[0]) or (row_n > p2[0] and row_n < p1[0]):
            distance = distance + 999999
    for col_n in empty_cols:
        if (col_n > p1[1] and col_n <p2[1]) or (col_n > p2[1] and col_n < p1[1]):
            distance = distance + 999999
    return distance
points = [p for p in np.argwhere(grid=="#")]
total_dist = 0
for i in range(len(points)):
    for j in range(i+1,len(points)):
        total_dist = total_dist + distance(points[i], points[j])
total_dist
```
