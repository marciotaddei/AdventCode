```python
# import re
import numpy as np
from collections import defaultdict
```

```python
with open("./input3.dat", "r") as f:
    lines = f.readlines()
lines[-1] = lines[-1]+"." #to avoid ragged

for line in lines:
    chars = line.split()
empty_row = np.array(list(['.' for _ in range(len(line)+1)]))
content_rows = np.array([['.']+[char for char in line] for line in lines])
input_array = np.vstack((empty_row, content_rows, empty_row))
del empty_row, content_rows
```

```python
sum = 0
for i in range(len(input_array)):
    number_string = ""
    for j in range(len(input_array[i])):
        if input_array[i,j].isnumeric():
            number_string = number_string + input_array[i,j]
        else:
            if len(number_string)>0:
                surroundings = {elem for elem in input_array[i-1,j-len(number_string)-1:j+1]}
                surroundings = surroundings.union({elem2 for elem2 in input_array[i+1,j-len(number_string)-1:j+1]})
                surroundings = surroundings.union({input_array[i,j-len(number_string)-1], input_array[i,j]})
                # print(number, surroundings)
                surroundings.discard('.')
                surroundings.discard('\n')
                if surroundings != set():
                    # print(number_string)
                    sum = sum + int(number_string)
            number_string = ""
sum
```

```python
sum = 0
geared_parts = []
gear_dict = defaultdict(list)
for i in range(len(input_array)):
    num_str = ""
    for j in range(len(input_array[i])):
        if input_array[i,j].isnumeric():
            num_str = num_str + input_array[i,j]
        else:
            if len(num_str)>0:
                surroundings = input_array[i-1:i+2, j-len(num_str)-1:j+1].copy()
                # print(surroundings)
                if '*' in surroundings:
                    gear_coords = tuple(map(tuple,np.argwhere(surroundings=='*') + (i-1,j-len(num_str)-1)-1))
                    # print(num_str, gear_coords)
                    # geared_parts = geared_parts + [gear_coords, int(num_str)]
                    gear_dict[gear_coords].append(int(num_str))
            num_str = ""
for key in gear_dict:
    if len(gear_dict[key])==2:
        sum = sum + gear_dict[key][0]*gear_dict[key][1]
sum
```
