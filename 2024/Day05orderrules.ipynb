```python
rules, updates = set(), []

with open("input05.txt", "r") as file:
    line = file.readline()
    while line != "\n":
        rules.add((int(line[0:2]), int(line[3:5])))
        line = file.readline()
    
    line = file.readline()
    while line != "":
        updates.append([int(x) for x in line.split(",")])
        line = file.readline()
```

```python
def check_order(update :list, rules :set):
    for i in range(1, len(update)):
        for rule in rules:
            if update[i]==rule[0]:
                if rule[1] in update[:i]:
                    return False
    return True
```

```python
correct_updates = [update for update in updates if check_order(update, rules)]
sum(l[len(l)//2] for l in correct_updates)
```

Part II

Possibility used: define a global before_dict with all the pages before a certain one (by direct rules only)<br>
Then, for each update, do the restriction of the before_dict to the pages involved<br>
Take a page with an empty (restricted) before and set it as the start of the update<br>
Recursively, call a function on an argument with that first one removed

```python
before = {num: set() for update in updates for num in update}
for rule in rules:
    before[rule[1]].add(rule[0])
```

```python
def order_series(series: list, before: dict):
    if len(series)==1:
        return series
    restricted_before = {pg:v.intersection(series) for pg,v in before.items() if pg in series}
    for pg in restricted_before:
        if restricted_before[pg] == set():
            return [pg] + order_series([x for x in series if x!=pg], restricted_before)
```

```python
def order_series(series: list, before: dict):
    if len(series)==1:
        return series
    restricted_before = {pg:v.intersection(series) for pg,v in before.items() if pg in series}
    for pg in restricted_before:
        if restricted_before[pg] == set():
            i = series.index(pg)
            return [pg] + order_series(series[:i]+series[i+1:], restricted_before)
```

```python
incorrect_updates = [update for update in updates if not check_order(update, rules)]
sum(order_series(l, before)[len(l)//2] for l in incorrect_updates)
```
