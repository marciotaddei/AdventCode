```python
def read_into_blocks(fname = "input09example.txt"):
    with open(fname, "r") as file:
        string = file.readline()

    blocks, id = [], 0
    for i in range(0,len(string)-(len(string)%2),2):
        blocks.append((id, int(string[i])) )
        blocks.append((".",int(string[i+1])))
        id +=1
    if (len(string)%2): blocks.append((id, int(string[-1])) )
    return [b for b in blocks if b!=(".", 0)]
```

```python
def defrag(blocks :list):
    while True:
        try:
            ind, n = next((i,no) for i, (id, no) in enumerate(blocks) if id == ".")
        except StopIteration: break

        last_id, last_amount = blocks[-1]
        if n < last_amount:
            blocks[ind]  = (last_id, n)
            blocks[-1] = (last_id, last_amount - n)
        elif n > last_amount:
            blocks[ind]  = (last_id, last_amount)
            blocks.insert(ind+1, (".", n-last_amount))
            blocks.pop(-1)
        else: #n == last_amount:
            blocks[ind]  = (last_id, last_amount)
            blocks.pop(-1)
            
        if blocks[-1][0] == ".": blocks.pop(-1)
    while blocks[-2][0] == blocks[-1][0]:
        blocks[-2] = (blocks[-2][0], blocks[-2][1]+ blocks[-1][1])
        blocks.pop(-1)
    return blocks
```

```python
def check_sum(blocks: list):
    sum, i = 0, 0
    for id, am in blocks:
        sum += id*(2*i+am-1)*am//2 #sum of arithm progr
        i   += am
    return sum
```

```python
blocks = read_into_blocks()
print(blocks)
blocks = defrag(blocks)
print(blocks)
check_sum(blocks)
```

```python
blocks = read_into_blocks("input09.txt")
print(blocks)
blocks = defrag(blocks)
print(blocks)
check_sum(blocks)
```

```python
def defrag_whole(blocks :list):
    blocks.append((".",1)) #added only for the joining of "." to call mov_idx+1 without IndexError
    last_id = blocks[-2][0] #last one not "."
    for mov_ID in range(last_id,0,-1): #first 0 in right place, stopped short
        mov_idx, mov_am = next((i, am) for i,(id, am) in enumerate(blocks) if id == mov_ID)
        if mov_idx <= next(i for i,(id, *_) in enumerate(blocks) if id == "."): break

        for idx, _, n in [(i, id, no) for i, (id, no) in enumerate(blocks) if id == "."]:
            if mov_idx < idx: break #only replace to the left, not to the right
            if n >= mov_am : #am for amount, the size of the block
                blocks[idx]     = (mov_ID, mov_am)
                blocks[mov_idx] = (".", mov_am)

                #joining some contiguous "."
                if blocks[mov_idx+1][0] == ".":
                    blocks[mov_idx] = (".", blocks[mov_idx][1] + blocks[mov_idx+1][1])
                    blocks = blocks[:mov_idx+1]+blocks[mov_idx+2:]
                if blocks[mov_idx-1][0] == ".":
                    blocks[mov_idx-1] = (".", blocks[mov_idx-1][1] + blocks[mov_idx][1])
                    blocks = blocks[:mov_idx]+blocks[mov_idx+1:]

                if n-mov_am: blocks.insert(idx+1, (".", n-mov_am))
                break

    if blocks[-1][0] == ".": blocks.pop(-1)
    return blocks
```

```python
def check_sum(blocks: list):
    sum, i = 0, 0
    for id, am in blocks:
        if id == ".": id=0
        sum += id*(2*i+am-1)*am//2 #sum of arithm progr
        i   += am
    return sum
```

```python
blocks = read_into_blocks()
print(blocks)
blocks = defrag_whole(blocks)
print(blocks)
check_sum(blocks)
```

```python
blocks = read_into_blocks("input09.txt")
print(blocks)
blocks = defrag_whole(blocks)
print(blocks)
check_sum(blocks)
```
