Part II: for 75 times, we need to count more cleverly.

Many repeated numbers, so do the calculation only once for each unique. Using multisets!
A cache helps slightly

```python
from multiset import Multiset as mset

input_ex = "125 17"
input = "4329 385 0 1444386 600463 19 1 56615"
```

```python
cache = {} # caching not so determinant for efficiency
def new_stones_from(stone):
    if stone in cache: return cache[stone]
    if stone == '0':
        cache[stone] = mset('1')
        return mset('1')
    if len(stone)%2 == 0:
        out = mset([stone[:len(stone)//2], stone[len(stone)//2:-1].lstrip("0")+stone[-1:]])
        cache[stone] = out
        return out
    cache[stone] = mset({str(2024*int(stone))})
    return mset({str(2024*int(stone))})

stones = mset(input.split())
for i in range(75):
    new_stones = mset()
    for st,count in stones.items():
        new_stones = new_stones.combine(mset([s for s in new_stones_from(st)])*count)
    stones = new_stones
    print(i, "\r", end="")
len(stones)
```
