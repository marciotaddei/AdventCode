```python
nodes, edges = set(), {}
with open("input23.txt", "r") as file:
    line = sorted(file.readline().strip("\n").split("-"))
    while line != [""]:
        if line[0] in edges: edges[line[0]].add(line[1])
        else: edges[line[0]] = {line[1]}
        # if line[1] in edges: edges[line[1]].add(line[0])
        # else: edges[line[1]] = {line[0]}
        # edges.add((line[0], line[1]))   
        nodes.add(line[0]); nodes.add(line[1])     
        line = sorted(file.readline().strip("\n").split("-"))
len(nodes), len(edges)
```

```python
n_triang, triangles = 0, set()
for n1 in edges:
    #catching triangle by first alphabetically
    if len(edges[n1]) <= 1: continue
    for n2 in edges[n1]:
        if n2 in edges:
            theintersect = edges[n1].intersection(edges[n2])
            if theintersect:
                triangles.update({(n1,n2,n3) for n3 in theintersect})
                # triangles += len(edges[n1].intersection(edges[n2]))
len(triangles), len({(n1,n2,n3) for (n1,n2,n3) in triangles if n1[0]=="t" or n2[0]=="t" or n3[0]=="t"})
```

Part II

```python
cliques, i = {3:triangles}, 3
edges.update({n: set() for n in nodes if n not in edges})

while cliques[i]:
    cliques[i+1] = set()
    for clique in cliques[i]:
        theintersect = nodes.copy()
        for n in clique:
            theintersect.intersection_update(edges[n])
        # print(theintersect)
        for n in theintersect:
            cliques[i+1].add(tuple(list(clique)+[n]))
    i+=1
print({k:cliques[k] for k in [max(cliques)-1, max(cliques)]})
```
