```python
import numpy as np
import matplotlib.pyplot as plt
```

Part One

```python
array = np.genfromtxt("input04.txt", dtype=str, delimiter=1)
little_dict = {"@":1, ".":0}
array = np.vectorize(little_dict.get)(array).astype(int)

movables = set()
for i,j in np.argwhere(array): #where array is one, not zero
    movable = np.sum(array[max(0,i-1):i+2, max(0,j-1):j+2])<5 #center + 4 more ones
    if movable: movables.add((int(i),int(j)))
array = 2*array
for i,j in movables: array[i,j]=1
plt.matshow(array)
len(movables)
```

Part Two

```python
array = np.genfromtxt("input04.txt", dtype=str, delimiter=1)
little_dict = {"@":1, ".":0}
array = np.vectorize(little_dict.get)(array).astype(int)
n_rolls_start = np.sum(array)

removing = True
while removing:
    removing = False
    for i,j in np.argwhere(array): #where array is one, not zero
        movable = np.sum(array[max(0,i-1):i+2, max(0,j-1):j+2])<5 #center + 4 more ones
        if movable: array[i,j]=0; removing=True

plt.matshow(array)
int(n_rolls_start - np.sum(array))
```
