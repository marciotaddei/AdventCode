```python
import numpy as np
the_array = np.loadtxt("./input6.dat", str)
the_array = the_array[:,1:].astype(int)
prod = 1
for col in range(the_array.shape[1]): #for each col
    t = the_array[0,col]
    n_sols = 0
    for j in range(1,t): #verify without t=0 or t=t (useless 0 distance)
        if j*(t-j)>the_array[1,col]:
            n_sols = n_sols+1
    # print(n_sols)
    prod = prod*n_sols
prod
```

```python
the_array = np.loadtxt("./input6.dat", str, delimiter=":")
time = int(the_array[0,1].replace(" ", ""))
dist = int(the_array[1,1].replace(" ", ""))

n_sols = 0
for j in range(1,time): #verify without t=0 or t=t (useless 0 distance)
##could also reduce to up to time/2, checking if even or odd for *2 or *2+1
    if j*(time-j)>dist:
        n_sols = n_sols+1
print(n_sols)
```
