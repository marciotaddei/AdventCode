```python
import numpy as np

garden = np.genfromtxt("input12.txt", dtype=str, delimiter=1)
garden
```

```python
def regions_from(garden: np.array) -> dict:
    region_per_letter = {}
    for l in 'ABCDEFGHIJKLMNOPQRSTUVWXYZ':
        region = np.argwhere(garden == l)
        if (region).any(): region_per_letter[l] = region

    regions = {}
    for l, region in region_per_letter.items():
        i, remain_reg = 0, {tuple(row) for row in region}
        while remain_reg:
            pos = remain_reg.pop()
            neighbors, new_neighbors = {pos}, {0}
            regions.update({(l, i): {pos}})
            while new_neighbors:
                new_neighbors = {tuple(np.array(n)+[0,1]) for n in neighbors}
                new_neighbors.update({tuple(np.array(n)+[0,-1]) for n in neighbors})
                new_neighbors.update({tuple(np.array(n)+[1,0]) for n in neighbors})
                new_neighbors.update({tuple(np.array(n)+[-1,0]) for n in neighbors})
                new_neighbors = {n for n in new_neighbors if n in remain_reg}
                regions[(l, i)].update(new_neighbors)
                remain_reg.difference_update(new_neighbors)
                neighbors = new_neighbors.copy()
            i+=1
    return regions

regions = regions_from(garden)
```

Part I

```python
region_data = {}
for key,region in regions.items():
    perimeter = 0
    for pos in region:
        neighbors = {tuple(np.array(pos)+[0,1]), tuple(np.array(pos)+[0,-1]), tuple(np.array(pos)+[1,0]), tuple(np.array(pos)+[-1,0])}
        perimeter += len(neighbors.difference(region))
    region_data[key] = (len(region), perimeter)

thesum = 0
for data in region_data.values():
    thesum += data[0]*data[1]
thesum    
```

```python
regions = regions_from(garden)

region_data = {}
for key,region in regions.items():
    sides = 0
    for pos in region:
        neighbors = [tuple(np.array(pos)+delta) in region for delta in [[-1,0], [0,-1], [0,1], [1,0]]]
        n_out = 4-sum(neighbors)
        if n_out == 4: sides = 4; break
        diagonals = [tuple(np.array(pos)+delta) in region for delta in [[-1,-1], [-1,1], [1,-1], [1,1]]]
        
        if n_out == 2:
            if not (neighbors[0] and neighbors[3]) and not (neighbors[1] and neighbors[2]) : sides += 1 #L shape
        elif n_out == 3: sides += 2

        #finding concave corners
        if not diagonals[0] and neighbors[0] and neighbors[1]: sides += 1
        if not diagonals[1] and neighbors[0] and neighbors[2]: sides += 1
        if not diagonals[2] and neighbors[1] and neighbors[3]: sides += 1
        if not diagonals[3] and neighbors[2] and neighbors[3]: sides += 1
        
    region_data[key] = (len(region), sides)
sum(data[0]*data[1] for data in region_data.values())#, region_data
```
