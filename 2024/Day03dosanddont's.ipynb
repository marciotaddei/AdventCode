```python
import re
def read_input(line):
    ct = 0
    l = re.split("mul\(",line)
    for string in l:
        try:
            string = string[:string.index(")")]
        except ValueError:
            continue
        if re.match("[0-9]{1,3},[0-9]{1,3}$", string):
            splitstring = string.split(",")
            ct += int(splitstring[0])*int(string.split(",")[1])
    return ct
```

```python
with open("./input03.txt", "r") as file:
    lines = file.readlines()
ct=0
for line in lines:
    ct += read_input(line)
ct
```

```python
import re
def read_input_enabled(line):
    ct = 0
    l = re.split("don\'t\(\)",line)
    ct += read_input(l[0])
    for substr in l[1:]:
        if "do()" not in substr: continue
        substr = substr[substr.index("do()")+4:]
        ct += read_input(substr)
    return ct
```

```python
with open("./input03.txt", "r") as file:
    lines = file.readlines()

ct = read_input_enabled("".join(lines)) #joined because some line ends in a don't(), the following one should begin disabled
ct
```
