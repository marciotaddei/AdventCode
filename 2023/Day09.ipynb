```python
import numpy as np
```

```python
filename = "input09.dat"
with open(filename, "r") as f:
    series = f.readlines()
sum = 0
for serie in series:
    serie = np.array(serie.split()).astype(int)
    # serie = np.array(series[i].split()).astype(int)
    aux_series = {}
    aux_series[0] = serie
    j = 0
    sum = sum + aux_series[0][-1]
    while set(aux_series[j][1:]-aux_series[j][:-1])!={0}:
        aux_series[j+1] = aux_series[j][1:]-aux_series[j][:-1]
        sum = sum + aux_series[j+1][-1]
        j = j+1
    # print(aux_series, sum)
sum
```
