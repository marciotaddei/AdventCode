```python
with open("input19.txt", "r") as file:
    patterns = file.readline().strip("\n").split(", ")
    file.readline()
    designs = [des.strip("\n") for des in file.readlines()]
# patterns, designs
```

```python
#explicitly building it as a DFS
def is_feasible(design):
    candidates = [design]
    while candidates:
        current = candidates.pop()
        new_cands = [current[len(pattern):] for pattern in patterns if current[:len(pattern)] == pattern]
        if "" in new_cands: return True
        candidates = candidates + new_cands
    return False
```

```python
ct=0
for design in designs:
    # print(is_feasible(design))
    ct += is_feasible(design)
ct
```

```python
with open("input19.txt", "r") as file:
    patterns = file.readline().strip("\n").split(", ")
    file.readline()
    designs = [des.strip("\n") for des in file.readlines()]
```

```python
#with multisets!
from multiset import Multiset as mset
def how_many_feas(design):
    # if not is_feasible(design): return 0
    candidates = mset({design})
    ct = 0
    while candidates:
        current, curr_multipl = candidates.popitem()
        for pattern in patterns:
            if current[:len(pattern)] == pattern:
                candidates = candidates.combine(mset({current[len(pattern):]:curr_multipl}))
        ct += candidates.pop("", 0)
    return ct
```

```python
ct=0
for design in designs:
    # print(how_many_feas(design))
    ct += how_many_feas(design)
ct
```

```python
#this one doesn't do a DFS (nor a BFS, it accesses sets in some arbitrary order, so a WFS (whatever-first search); takes much longer for it)
def is_feasible(design):
    candidates = {design}
    while candidates:
        current = candidates.pop()
        new_cands = {current[len(pattern):] for pattern in patterns if current[:len(pattern)] == pattern}
        if "" in new_cands: return True
        candidates.update(new_cands)
    return False
```

```python
#way too slow. Maybe because it's making a BFS
def is_feasible(design):
    if design == "": return True
    reduced = set()
    for pattern in patterns:
        if design[:len(pattern)] == pattern:
            reduced.add(design[len(pattern):])
    return True in {is_feasible(red_des) for red_des in reduced}
```

```python
#this very slow for Part II
def how_many_feas(design):
    # if not is_feasible(design): return 0
    candidates = [design]
    ct = 0
    while candidates:
        current = candidates.pop()
        new_cands = [current[len(pattern):] for pattern in patterns if current[:len(pattern)] == pattern]
        candidates = candidates + new_cands
        ct += new_cands.count("")
    return ct
```
