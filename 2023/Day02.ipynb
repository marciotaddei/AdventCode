```python
import re
```

```python
with open("input2.dat", "r") as f:
    lines = f.readlines()
sum=0
for line in lines:
    # print(line)
    line = re.sub("[:,;]",  "",  line)
    line = line.split()
    info = zip(line[2::2], line[3::2])
    for num, col in info:
        if int(num)>14:
            break
        if int(num)==14 and col != "blue":
            break
        if int(num)==13 and col == "red":
            break
    else:
        sum = sum + int(line[1])
sum
```

```python
with open("./input2.dat", "r") as f:
    lines = f.readlines()
sum=0
for line in lines:
    # print(line)
    line = re.sub("[:,;]",  "",  line).split()
    info = zip(line[2::2], line[3::2])
    # min_r, min_g, min_b = 0, 0, 0
    min_dict = {'red': 0, 'green': 0, 'blue': 0}
    for num, color in info:
        num = int(num)
        if min_dict[color] < num:
            min_dict[color] = num
    sum = sum + min_dict['red']*min_dict['green']*min_dict['blue']
sum
```
