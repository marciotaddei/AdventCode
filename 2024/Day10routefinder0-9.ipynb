```python
import numpy as np
np.set_printoptions(linewidth=1000, threshold=1000)
```

```python
topography = np.genfromtxt('input10.txt', dtype=int, delimiter=1)
topography = np.pad(topography, 1, mode='constant', constant_values=-1)

ct = 0 
head_candidates = np.argwhere(topography == 0)
for trailhead in head_candidates:
    pos, h, trailpos = trailhead, 0, np.array([trailhead])
    for h in range(1,10):
        new_trailpos = []
        for pos in trailpos:
            if topography[pos[0]+1, pos[1]] == h:
                new_trailpos.append((pos[0]+1, pos[1]))
            if topography[pos[0]-1, pos[1]] == h:
                new_trailpos.append((pos[0]-1, pos[1]))
            if topography[pos[0], pos[1]+1] == h:
                new_trailpos.append((pos[0], pos[1]+1))
            if topography[pos[0], pos[1]-1] == h:
                new_trailpos.append((pos[0], pos[1]-1))
        if not new_trailpos: break
        trailpos = np.array(new_trailpos)
    ct += np.unique(trailpos, axis=0).shape[0]
ct
```
