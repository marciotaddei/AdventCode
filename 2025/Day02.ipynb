Part Two

```python
ranges_IDs = []
for pair in prob_input.split(","):
    (n1,n2) = (int(n) for n in pair.split("-"))
    #only numbers with even amount of digits allowed
    digits_n1, digits_n2 = math.floor(math.log10(n1))+1, math.floor(math.log10(n2))+1
    if digits_n1==digits_n2:
        ranges_IDs.append((n1,n2))
    else: #only one extra digit in n2
        ranges_IDs.append((n1,10**digits_n1-1))
        ranges_IDs.append((10**(digits_n2-1),n2))

invalid_IDs = set()
for n1,n2 in ranges_IDs:
    digits_n = math.floor(math.log10(n1))+1
    print(f"Interval {n1}-{n2}, with {digits_n}-digit numbers")
    for j in range(2,digits_n+1):
        # if j non-prime:continue (no need, too fast anyway)
        if digits_n%j==0:
            factor = sum(10**(i*digits_n//j) for i in range(j)) #111, 10101, 1001001, correspondingly, with j ones and as many zeroes as to fit the full number
            start = -(-n1//factor)*factor
            print(f"\t{j}: multiples of {factor} starting at {start}")
            for invID in range(start, n2+1, factor):
                invalid_IDs.add(invID)

print(f"\nThe sum of invalid IDs is {sum(invalid_IDs)}")
invalid_IDs
```
