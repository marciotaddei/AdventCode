```python
def hasher(string):
    value = 0
    for ch in string:
        value += ord(ch)
        value *= 17
        value = value%256
    return value
```

```python
with open("input15.dat", "r") as f:
    init_seq = f.readline().split(',')

the_sum = 0
for step in init_seq:
    the_sum += hasher(step)
the_sum
```

```python
def HASHMAP(string):
    label = string.strip("=-0123456789")
    box = boxes[hasher(label)]
    if string[len(label)] == "-":
        box.pop(label, None)
    elif string[len(label)] == "=":
        box.update({label: int(string[len(label)+1:])})
    else:
        raise ValueError("Character should've been = or - here")
    return None
```

```python
with open("input15.dat", "r") as f:
    init_seq = f.readline().split(',')

boxes = [{} for _ in range(256)]
# boxes = [{}.copy]*256 #why doesn't this work? All dict changed at once
set(HASHMAP(string) for string in init_seq) #HASHMAP alters boxes directly, return None
f_power = 0
for i in range(len(boxes)):
    focal_l_iter = iter(boxes[i].values())
    for j in range(len(boxes[i])):
        f_power += (i+1)*(j+1)*next(focal_l_iter)
f_power
```
