```python
full_array = np.loadtxt("./input4.dat", str)
# colon_pos = 2 
vertbar_pos = np.argwhere(full_array[0]=="|")[0,0]
winning_nums = full_array[:,2:vertbar_pos].astype(int)
your_nums    = full_array[:,vertbar_pos+1:].astype(int)

card_copies = np.ones((full_array.shape[0],1),int)

for i in range(full_array.shape[0]):
    #intersect = set(your_nums[i]).intersection(set(winning_nums[i])) #intersection
    #n_matches_card = len(intersect)
    n_matches_card = len(set(your_nums[i]).intersection(set(winning_nums[i])))
    card_copies[i+1:i+n_matches_card+1] = card_copies[i+1:i+n_matches_card+1] + card_copies[i]
    # print(intersect)
card_copies.sum()
```
